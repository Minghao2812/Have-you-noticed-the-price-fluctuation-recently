It has been 16 months since the initial action of in the Section 301 investigation by the US Trade Representative, and we could feel the impact of the trade war between two **giant economic entity** not only in media, but also in our daily life via goods' price. To make it simple, economists often choose soybean, crude oil and iron's price as three indicators stands for agriculture, energy and metal, which consist of a country's lifeline.

## Timetable
In order to monitor the price trend, to compare it against trade war timetable is not a bad idea. We could split the price trend into pieces using time spots to special issues.

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker
import matplotlib.axis as ax
```

```python
path = './'
file_soybean = 'soybean.csv'
file_crude_oil = 'crude_oil.csv'
file_iron = 'iron.csv'

'''
file_gold = 'gold.csv'
file_aluminum = 'aluminum.csv'
file_rubber = 'rubber.csv'
file_gdp = 'GDP_PPP.csv'
'''
```

```python
df_soybean = pd.read_csv(path+file_soybean) #soybean - food
df_crude_oil = pd.read_csv(path+file_crude_oil) #oil - energy
df_iron = pd.read_csv(path+file_iron) #iron - metal

'''
df_gold = pd.read_csv(path+file_gold) #gold - measurement
df_aluminum = pd.read_csv(path+file_aluminum)
df_rubber = pd.read_csv(path+file_rubber)
df_gdp = pd.read_csv(path+file_gdp)
'''
```

So, I collected two lists of time. The first one contains time that there are some issues happen between China and the US. Not all time spots have negative issues; there are negotiations as well. The second one contains time spots all about tariff increases. Both the US and China has increased tariff three times, which sums as 6 times altogether,and the US always goes first. However, due to the lack of data, I just collected 5 time spots and the data of third increase by China is not available yet. Here are the lists:

```python
trade_war_list = ['Mar-17','Apr-17','Jul-17','Aug-17','Jan-18','Mar-18','Apr-18','Jun-18','Jul-18','Aug-18','Sept-18','Dec-18','Feb-19','May-19','Jun-19','Aug-19']
tariff_list = ['Jul-18','Aug-18','Sep-18','May-19','Sep-19']
print('The first list: all important time spots.')
print(trade_war_list,'\n')
print('The second list: about tariff increases.')
print(tariff_list)
```

## Price Trend
Then I drew three plots about price trend during Jan 2017 to Sep 2019. In each plot, I marked those time spots as well. **Blue lines** are general important time spots, and **red lines** are time spots when tariff increased, which are included by the former ones. Let's look at them separately. 

As for soybean, we can almost say that every time there is an issue, there is a price fluctuation. Especially in Sep 2018 and May 2019, there are two duties aim at soybean. Contrasting to the fluctuation of soybean's price, crude oil and iron's prices are less correlated to the duty however. That means there are much fluctuation without duty imposing, and duty matters not that much to their prices.

To look down from above, soybean is more expensive while crude oil and iron are less expensive. Those are not good signs. That means many farmers may suffer, while people in cities may pay more to afford transportation and housing.

```python
soybean_time = df_soybean['time']
soybean_price = df_soybean['price']

crude_oil_time = df_crude_oil['time']
crude_oil_price = df_crude_oil['price']

iron_time = df_iron['time']
iron_price = df_iron['price']
```

```python
plt.plot(soybean_time, soybean_price,'go-')
plt.title('Soybean\'s Price During Trade War')
plt.xlabel('Time')
plt.ylabel('Money/$')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(4))

for time in trade_war_list:
    plt.axvline(time, color='b', linestyle='--', linewidth=1)
for time in tariff_list:
    plt.axvline(time, color='r', linestyle='--', linewidth=1)

ax = plt.gca()
ax.spines['top'].set_visible(False) #erase top frame
#ax.spines['bottom'].set_visible(False) #erase bottom frame
ax.spines['left'].set_visible(False) #erase left frame
ax.spines['right'].set_visible(False) #erase fight frame
plt.show()
```

```python
plt.plot(crude_oil_time, crude_oil_price,'go-')
plt.title('Crude Oil\'s Price During Trade War')
plt.xlabel('Time')
plt.ylabel('Money/$')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(4))

for time in trade_war_list:
    plt.axvline(time, color='b', linestyle='--', linewidth=1)
for time in tariff_list:
    plt.axvline(time, color='r', linestyle='--', linewidth=1)
    
ax = plt.gca()
ax.spines['top'].set_visible(False) #erase top frame
#ax.spines['bottom'].set_visible(False) #erase bottom frame
ax.spines['left'].set_visible(False) #erase left frame
ax.spines['right'].set_visible(False) #erase fight frame
plt.show()
```

```python
plt.plot(iron_time, iron_price,'go-')
plt.title('Iron\'s Price During Trade War')
plt.xlabel('Time')
plt.ylabel('Money/$')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(4))

for time in trade_war_list:
    plt.axvline(time, color='b', linestyle='--', linewidth=1)
for time in tariff_list:
    plt.axvline(time, color='r', linestyle='--', linewidth=1)
    
ax = plt.gca()
ax.spines['top'].set_visible(False) #erase top frame
#ax.spines['bottom'].set_visible(False) #erase bottom frame
ax.spines['left'].set_visible(False) #erase left frame
ax.spines['right'].set_visible(False) #erase fight frame
plt.show()
```

```python
soybean_std = np.std(soybean_price)
crude_oil_std = np.std(crude_oil_price)
iron_std = np.std(iron_price)
dict_std = {'cata':['Soybean Std','Crude Oil Std','Iron Std'],'std':[soybean_std,crude_oil_std,iron_std]}
df_std = pd.DataFrame(dict_std)


#dict_price = {'Soybean Price':soybean_price,'Crude Oil Price':crude_oil_price,'Iron Price':iron_price}
#df_price = pd.DataFrame(dict_price)

#std = [soybean_std,crude_oil_std,iron_std]
#cata = ['Soybean Std','Crude Oil Std','Iron Std']
#std
```

## The Standard Deviation of Three Goods
From the bar chart of standard deviation, we could easily find that soybean's price is most unstable: its standard deviation value is twice to that of crude oil price. Different to goods in energy of metal market, agricultural goods are essentially unstable because of many natural factors such as flood, drought and snowstorm.

```python
plt.figure(figsize=[4,5])
plt.title('Standard Deviation')
plt.ylabel('Money/$')
plt.grid(axis = 'y',c = 'w')
ax = plt.gca()
ax.spines['top'].set_visible(False) #erase top frame
ax.spines['bottom'].set_visible(False) #erase bottom frame
ax.spines['left'].set_visible(False) #erase left frame
ax.spines['right'].set_visible(False) #erase fight frame

plt.bar(df_std['cata'],df_std['std'])
plt.show() # this bar chart is compact :)
```
